## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here
sns.scatterplot(data=penguins, x='bill_length_mm', \
                y='bill_depth_mm', hue="species")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)");
```

**Your interpretation (2–3 sentences):**
From the scatter plot it seems as though Adelie Penguins have more Bill Depth but less Bill Length on average. On the reverse of this, the Gentoo penguins have more bill length but smaller bill depth. The Chinstrap penguins have a good medium between the length and depth.

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
# TODO: Histogram with 10 bins

blength = penguins['bill_length_mm']
plt.hist(blength, bins = 10, density = True)
plt.title("Bill Length (10 bins)")
plt.xlabel("bill_length_mm")
plt.ylabel("Density")
```

```python
# TODO: Histogram with 30 bins
blength = penguins['bill_length_mm']
plt.hist(blength, bins = 30, density = True)
plt.title("Bill Length (30 bins)")
plt.xlabel("bill_length_mm")
plt.ylabel("Density")
```

**Your comparison (1–2 sentences):**

When the number of bins is changed, more bars are added to the histogram to show more data.

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
# TODO: Boxplot of body mass by species
sns.boxplot(data=penguins, y="body_mass_g", hue="species")
plt.title("Body Mass by Species");
```

**Your summary (1–2 sentences):**

From the box plot, it appears that Gentoo are the heaviest on average, with Adelie and Chinstrap being roughly equally on average however with Adelie have more variation on the lower and upper ends.
